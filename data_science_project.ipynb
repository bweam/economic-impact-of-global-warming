#Sea level

URL : https://drive.google.com/file/d/1HH136d0HMk7UOc6SNZvZpZrnZK8-6I5k/view?usp=sharing

```
import math
import seaborn as sbs
import matplotlib.pyplot as pyplot
import pandas as pd
seaLevel = pd.read_csv('sea_level.csv',skiprows=5)
def get_mean(data):
  listedData = (data.to_numpy().tolist())
  for n in range(len(listedData)):
    filteredArr = list(filter(lambda x: x!=0 and (not math.isnan(x)),listedData[n]))
    if len(filteredArr)!=0:
      summation = sum(filteredArr)/len(filteredArr)
    else:
      summation=0;
    listedData[n] = summation
  return listedData
seaLevel['year'] = seaLevel['year'].astype(int)
seaLevel['TOPEX/Poseidon'] = seaLevel['TOPEX/Poseidon'].fillna(0)
seaLevel['Sea level anomaly'] = get_mean(seaLevel[seaLevel.columns[1:]])
fig, ax = pyplot.subplots(figsize = (20,20))
formattedDf = seaLevel.groupby('year').mean()
seaPlot = sbs.lineplot(data=formattedDf,x='year',y='Sea level anomaly',linewidth=4)
seaPlot.set_title('Global Mean Sea Level Anomaly', fontsize=30)
seaPlot.set_yticklabels(seaPlot.get_yticks(),size=20)
ax.set_xlabel('Year',fontsize=30)
ax.set_ylabel('Sea Level Anomaly',fontsize=30)
seaPlot.set_xticklabels(seaPlot.get_xticks(),size=20)

```
